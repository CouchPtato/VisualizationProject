```python
import numpy as np

# a. Create a 2D array with random values from 0 to 1 and calculate mean, std, and variance along the second axis.
ARR1 = np.random.rand(4, 5)  # Example 4x5 array
mean_arr1 = np.mean(ARR1, axis=1)
std_arr1 = np.std(ARR1, axis=1)
var_arr1 = np.var(ARR1, axis=1)
print("Mean:", mean_arr1)
print("Standard Deviation:", std_arr1)
print("Variance:", var_arr1, "\n")

# b. Create a 2D array of size m x n, reshape to n x m and print properties
m, n = 4, 3
ARR2 = np.random.randint(0, 10, (m, n))
print("Shape:", ARR2.shape, "Type:", type(ARR2), "Data Type:", ARR2.dtype)
ARR2_reshaped = ARR2.reshape(n, m)
print("Reshaped Array:\n", ARR2_reshaped, "\n")

# c. Check for zero, non-zero, and NaN elements in a 1D array
ARR3 = np.array([0, 1, np.nan, 3, 0, 5, np.nan])
zero_indices = np.where(ARR3 == 0)[0]
non_zero_indices = np.where(ARR3 != 0)[0]
nan_indices = np.where(np.isnan(ARR3))[0]
print("Zero Indices:", zero_indices)
print("Non-zero Indices:", non_zero_indices)
print("NaN Indices:", nan_indices, "\n")

# d. Create random arrays, perform calculations, and find covariance and correlation
Array1, Array2, Array3 = np.random.rand(5), np.random.rand(5), np.random.rand(5)
Array4 = Array3 - Array2
Array5 = 2 * Array1
covariance = np.cov(Array1, Array4)[0, 1]
correlation = np.corrcoef(Array1, Array5)[0, 1]
print("Covariance (Array1, Array4):", covariance)
print("Correlation (Array1, Array5):", correlation, "\n")

# e. Sum of the first half and product of the second half of two arrays
Array1, Array2 = np.random.rand(10), np.random.rand(10)
sum_first_half = np.sum(Array1[:5] + Array2[:5])
product_second_half = np.prod(Array1[5:] * Array2[5:])
print("Sum of first half:", sum_first_half)
print("Product of second half:", product_second_half, "\n")

# f. Determine the memory size occupied by an array
random_array = np.random.rand(10, 10)
print("Memory Size of array:", random_array.nbytes, "bytes", "\n")

# g. Swap rows, reverse column, and store in another array
m, n = 4, 5
ARR4 = np.random.randint(10, 100, (m, n))
ARR4_swapped = ARR4.copy()
ARR4_swapped[[0, 1]] = ARR4_swapped[[1, 0]]  # Swap rows 0 and 1
ARR4_swapped[:, 2] = ARR4_swapped[:, 2][::-1]  # Reverse column 2
print("Updated Array:\n", ARR4_swapped, "\n")
```

```python
import pandas as pd

# a. Create a series, display sorted by index and values
series_a = pd.Series([5, 3, 8, 6, 7])
print("Sorted by index:\n", series_a.sort_index())
print("Sorted by values:\n", series_a.sort_values(), "\n")

# b. Create a series with duplicate values, find min and max ranks
series_b = pd.Series([4, 2, 4, 6, 1, 1, 6])
print("Ranks (method='first'):\n", series_b.rank(method='first'))
print("Ranks (method='max'):\n", series_b.rank(method='max'), "\n")

# c. Display index of min and max element
print("Index of Min:", series_a.idxmin())
print("Index of Max:", series_a.idxmax(), "\n")
```

```python
# Generating DataFrame with 3 columns, 50 rows
df = pd.DataFrame(np.random.rand(50, 3), columns=['A', 'B', 'C'])
# Introducing 10% null values
null_indices = np.random.choice(df.index, size=int(len(df) * 0.1), replace=False)
df.loc[null_indices, 'A'] = np.nan

# a. Count missing values
print("Missing values count:\n", df.isnull().sum(), "\n")

# b. Drop columns with more than 5 null values
df_cleaned = df.dropna(axis=1, thresh=45)
print("Data after dropping columns with >5 null values:\n", df_cleaned, "\n")

# c. Drop row with max sum
row_max_sum = df.sum(axis=1).idxmax()
df_dropped = df.drop(index=row_max_sum)
print("Data after dropping row with max sum:\n", df_dropped, "\n")

# d. Sort DataFrame on the first column
df_sorted = df.sort_values(by='A')
print("Sorted DataFrame:\n", df_sorted, "\n")

# e. Remove duplicates from the first column
df_unique = df.drop_duplicates(subset='A')
print("DataFrame after removing duplicates from column A:\n", df_unique, "\n")

# f. Find correlation and covariance
print("Correlation between A and B:", df['A'].corr(df['B']))
print("Covariance between B and C:", df['B'].cov(df['C']), "\n")

# g. Discretize the second column into 5 bins
df['B_bins'] = pd.cut(df['B'], bins=5)
print("DataFrame with discretized B column:\n", df[['B', 'B_bins']], "\n")
```

```python
import pandas as pd
import numpy as np

# Define data for workshop 1
names_workshop1 = ['John', 'Alice', 'Bob', 'Clara', 'John', 'Alice', 'David', 'Eve', 'Bob', 'Clara']
dates_workshop1 = pd.date_range(start='2024-11-01', periods=5, freq='D').tolist() * 2  # Repeat for two rounds
durations_workshop1 = np.random.choice([30, 40, 50], size=10)  # Random durations (30, 40, 50)

# Create DataFrame for workshop 1
df1 = pd.DataFrame({
    'Name': names_workshop1,
    'Date': dates_workshop1,
    'Duration (minutes)': durations_workshop1
})

# Save DataFrame to Excel
df1.to_excel('workshop1.xlsx', index=False)

# Define data for workshop 2
names_workshop2 = ['Sara', 'Eve', 'John', 'Alice', 'Clara', 'David', 'Sara', 'Eve', 'John', 'David']
dates_workshop2 = pd.date_range(start='2024-11-01', periods=5, freq='D').tolist() * 2  # Repeat for two rounds
durations_workshop2 = np.random.choice([30, 40, 50], size=10)  # Random durations (30, 40, 50)

# Create DataFrame for workshop 2
df2 = pd.DataFrame({
    'Name': names_workshop2,
    'Date': dates_workshop2,
    'Duration (minutes)': durations_workshop2
})

# Save DataFrame to Excel
df2.to_excel('workshop2.xlsx', index=False)

print("Excel files 'workshop1.xlsx' and 'workshop2.xlsx' have been created.")
```

```python
import pandas as pd

# Read the data from the Excel files
df1 = pd.read_excel('workshop1.xlsx')
df2 = pd.read_excel('workshop2.xlsx')

# a. Perform merging of the two data frames to find the names of students who had attended both workshops
merged_df = pd.merge(df1, df2, on='Name', how='inner')  # Merge on 'Name' column, keep only common attendees
print("Students who attended both workshops:")
print(merged_df[['Name']].drop_duplicates())

# b. Find names of all students who have attended a single workshop only
only_workshop1 = df1[~df1['Name'].isin(df2['Name'])]
only_workshop2 = df2[~df2['Name'].isin(df1['Name'])]
single_workshop_attendees = pd.concat([only_workshop1, only_workshop2])
print("\nStudents who attended only one workshop:")
print(single_workshop_attendees[['Name']].drop_duplicates())

# c. Merge two data frames row-wise and find the total number of records in the data frame
df_combined = pd.concat([df1, df2], ignore_index=True)  # Concatenate the dataframes row-wise
print("\nTotal number of records after row-wise merge:", len(df_combined))

# d. Merge two data frames row-wise and use 'Name' and 'Date' as multi-row indexes
df_combined_multi_index = pd.concat([df1, df2], ignore_index=True)
df_combined_multi_index.set_index(['Name', 'Date'], inplace=True)
print("\nData frame with multi-row indexes:")
print(df_combined_multi_index.head())

# Generate descriptive statistics for this hierarchical data frame
print("\nDescriptive Statistics for the multi-indexed data frame:")
print(df_combined_multi_index.describe())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy import stats

# Load the Iris dataset and create a DataFrame
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# a. Display data types of the dataset
print("Data Info:")
print(df_iris.info(), "\n")

# b. Check for missing values
print("\nMissing Values:")
print(df_iris.isnull().sum(), "\n")

# c. Plot a bar chart to show the frequency of each class label in the data
plt.figure(figsize=(8, 5))
sns.countplot(data=df_iris, x='species')
plt.title('Frequency of Each Species')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()

# d. Scatter plot for Petal Length vs Sepal Length with regression line
plt.figure(figsize=(8, 6))
sns.regplot(data=df_iris, x='petal length (cm)', y='sepal length (cm)', scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('Petal Length vs Sepal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.show()

# e. Density plot for Petal Width
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df_iris, x='petal width (cm)', hue='species', fill=True)
plt.title('Density Distribution for Petal Width')
plt.xlabel('Petal Width (cm)')
plt.show()

# f. Pair plot to show pairwise bivariate distribution in the Iris Dataset
sns.pairplot(df_iris, hue='species', diag_kind='kde', palette='Set2')
plt.suptitle('Pairwise Bivariate Distribution in Iris Dataset', y=1.02)
plt.show()

# g. Heatmap for correlation between two numeric attributes
plt.figure(figsize=(6, 5))
sns.heatmap(df_iris[['sepal length (cm)', 'petal length (cm)']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap between Sepal Length and Petal Length')
plt.show()

# h. Statistical computations: mean, mode, median, std dev, confidence interval, standard error
numeric_features = df_iris.select_dtypes(include=[np.number]).columns

stats_df = pd.DataFrame(index=['mean', 'median', 'mode', 'std_dev', 'conf_int', 'std_error'])
for feature in numeric_features:
    data = df_iris[feature]
    mean = data.mean()
    median = data.median()
    mode = data.mode()[0]
    std_dev = data.std()
    conf_int = stats.norm.interval(0.95, loc=mean, scale=std_dev/np.sqrt(len(data)))
    std_error = std_dev / np.sqrt(len(data))
    
    stats_df[feature] = [mean, median, mode, std_dev, conf_int, std_error]

print("\nStatistical Summary:")
print(stats_df, "\n")

# i. Correlation coefficients and heatmap for all numeric features
plt.figure(figsize=(8, 6))
correlation_matrix = df_iris.iloc[:, :-1].corr()  # Exclude the species column for correlations
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap for Numeric Features')
plt.show()
```

```python
data = pd.read_csv(r'C:\Users\singh\OneDrive\Desktop\VisualizationProject\Practicals\Titanic-Dataset.csv')

df = pd.DataFrame(data)

# a. Clean the data by dropping the column which has the largest number of missing values.
# 'Cabin' has the most missing values, so we drop it.
df_cleaned = df.drop(columns=['Cabin'])

# b. Find total number of passengers with age more than 30
passengers_above_30 = df_cleaned[df_cleaned['Age'] > 30].shape[0]
print(f"Total number of passengers with age more than 30: {passengers_above_30}")

# c. Find total fare paid by passengers of second class
fare_second_class = df_cleaned[df_cleaned['Pclass'] == 2]['Fare'].sum()
print(f"Total fare paid by passengers of second class: {fare_second_class}")

# d. Compare number of survivors of each passenger class
survivors_by_class = df_cleaned.groupby('Pclass')['Survived'].sum()
print("\nNumber of survivors in each passenger class:")
print(survivors_by_class)

# e. Compute descriptive statistics for age attribute gender wise
age_statistics_gender = df_cleaned.groupby('Sex')['Age'].describe()
print("\nDescriptive statistics for Age attribute by gender:")
print(age_statistics_gender)

# f. Draw a scatter plot for passenger fare paid by Female and Male passengers separately
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Fare', y='Age', hue='Sex', data=df_cleaned)
plt.title('Scatter Plot of Fare vs Age by Gender')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend(title='Gender')
plt.show()

# g. Compare density distribution for features age and passenger fare
plt.figure(figsize=(8, 5))
sns.kdeplot(df_cleaned['Age'].dropna(), label='Age', fill=True)
sns.kdeplot(df_cleaned['Fare'], label='Fare', fill=True)
plt.title('Density Distribution of Age and Fare')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# h. Draw the pie chart for three groups labelled as class 1, class 2, class 3
class_counts = df_cleaned['Pclass'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=['Class 1', 'Class 2', 'Class 3'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Passenger Class Distribution')
plt.show()

# i. Find % of survived passengers for each class and answer the question Did class play a role in survival?
survival_rate_by_class = df_cleaned.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival rate by passenger class:")
print(survival_rate_by_class)
```

```python
# Given data as a DataFrame
data = {
    'FamilyName': ['Shah', 'Vats', 'Vats', 'Kumar', 'Vats', 'Kumar', 'Shah', 'Shah', 'Kumar', 'Vats'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
    'MonthlyIncome': [44000, 65000, 43150, 66500, 255000, 103000, 55000, 112400, 81030, 71900]
}
df_family = pd.DataFrame(data)

# a. Family-wise gross monthly income
gross_income = df_family.groupby('FamilyName')['MonthlyIncome'].sum()
print("Family-wise Gross Monthly Income:\n", gross_income, "\n")

# b. Highest and lowest monthly income for each family
print("Highest Income:\n", df_family.groupby('FamilyName')['MonthlyIncome'].max())
print("Lowest Income:\n", df_family.groupby('FamilyName')['MonthlyIncome'].min(), "\n")

# c. Monthly income below 80,000
below_80000 = df_family[df_family['MonthlyIncome'] < 80000]
print("Income below 80,000:\n", below_80000, "\n")

# d. Total number of females with average monthly income
female_avg_income = df_family[df_family['Gender'] == 'Female']
print("Number of Females:", len(female_avg_income), "Average Income:", female_avg_income['MonthlyIncome'].mean(), "\n")

# e. Delete rows with income less than average
average_income = df_family['MonthlyIncome'].mean()
df_filtered = df_family[df_family['MonthlyIncome'] >= average_income]
print("Filtered Data:\n", df_filtered)
```
